```python
import fasttext
import hyperopt
from hyperopt import fmin, tpe, Trials

# Define the hyperparameter space to search over
space = {
    'lr': hyperopt.hp.loguniform('lr', -5, 0),
    'epoch': hyperopt.hp.quniform('epoch', 5, 50, 1),
    'wordNgrams': hyperopt.hp.quniform('wordNgrams', 1, 10, 1),
    'minCount': hyperopt.hp.quniform('minCount', 1, 100, 1),
    'bucket': hyperopt.hp.quniform('bucket', 100000, 1000000, 10000),
    'neg': hyperopt.hp.quniform('neg', 5, 30, 1),
    'ws': hyperopt.hp.quniform('ws', 2, 10, 1),
    't': hyperopt.hp.quniform('t', 0.00001, 0.1, 0.00001),
    'lrUpdateRate': hyperopt.hp.quniform('lrUpdateRate', 5, 100, 1),
}


def objective(params):
    global model
    model = fasttext.train_supervised(
        input="train_data.txt",
        lr=params['lr'],
        epoch=int(params['epoch']),
        loss='ova',
        wordNgrams=int(params['wordNgrams']),
        dim=30,
        thread=24,
        verbose=100
    )
    # Evaluate the model on a validation set and return the loss
    test_result = model.test("test_data.txt")
    loss = 1 - test_result[1] # Using error rate instead of accuracy as the objective function
    return {'loss': loss, 'status': hyperopt.STATUS_OK, 'test_accuracy': test_result[1]}

# Run the optimization using TPE algorithm
# Run the optimization using BOTPE algorithm
trials = Trials()
best_model = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=100,
            trials=trials)

print("Best hyperparameters found:", best_model)
print("Best test accuracy:", 1 - trials.best_trial['result']['loss'])



```

```python
def split_file(input_file, output_file1, output_file2, num_lines):
    # Open the input file
    with open(input_file, 'r') as f:
        # Read all the lines
        lines = f.readlines()
        
        # Split the lines into two lists
        lines1 = lines[:num_lines]
        lines2 = lines[num_lines:]
        
        # Write the lines to output files
        with open(output_file1, 'w') as f1:
            f1.writelines(lines1)
        with open(output_file2, 'w') as f2:
            f2.writelines(lines2)
```
